```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataset
data = {
    'District': ['Kambia', 'Kailahun', 'Pujehun'],
    'Males': [398, 153, 279],
    'Females': [207, 84, 146]
}

df = pd.DataFrame(data)

# Melt the DataFrame for easier plotting with seaborn
df_melted = df.melt(id_vars='District', var_name='Gender', value_name='Count')

# Calculate percentages for labels
df_melted['Percentage'] = (df_melted['Count'] / df_melted.groupby('District')['Count'].transform('sum')) * 100

# Set the color palette
sns.set_palette("bright")

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='District', y='Count', hue='Gender', data=df_melted)

# Add labels on top of bars
for p in bar_plot.patches:
    count_value = int(p.get_height())
    if count_value > 0:  # Only annotate if the height is greater than 0
        bar_plot.annotate(f'{count_value} ({count_value / df_melted['Count'].sum() * 100:.1f}%)',
                          (p.get_x() + p.get_width() / 2., p.get_height()),
                          ha='center', va='baseline',
                          fontsize=10,
                          color='black',
                          rotation=0,
                          xytext=(0, 5),  # 5 points vertical offset
                          textcoords='offset points')

# Customize the plot
plt.title('Population Distribution by Gender in Districts')
plt.xlabel('District')
plt.ylabel('Number of Individuals')
plt.legend(title='Gender')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataset
data = {
    'Output': ['Output 3', 'Output 4'],
    'Males': [372, 437],
    'Females': [162, 296]
}

df = pd.DataFrame(data)

# Melt the DataFrame for easier plotting with seaborn
df_melted = df.melt(id_vars='Output', var_name='Gender', value_name='Count')

# Calculate percentages for labels
df_melted['Percentage'] = (df_melted['Count'] / df_melted.groupby('Output')['Count'].transform('sum')) * 100

# Set the color palette
sns.set_palette("bright")

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Output', y='Count', hue='Gender', data=df_melted)

# Add labels on top of bars, omitting 0 (0%)
for p in bar_plot.patches:
    height = p.get_height()
    if height > 0:  # Only label if the height is greater than 0
        bar_plot.annotate(f'{int(height)} ({height / df_melted['Count'].sum() * 100:.1f}%)',
                          (p.get_x() + p.get_width() / 2., height),
                          ha='center', va='baseline',
                          fontsize=10,
                          color='black',
                          rotation=0,
                          xytext=(0, 5),  # 5 points vertical offset
                          textcoords='offset points')

# Customize the plot
plt.title('Population Distribution by Gender in Outputs')
plt.xlabel('Output')
plt.ylabel('Number of Individuals')
plt.legend(title='Gender')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataset
data = {
    'District': ['Kambia', 'Kailahun', 'Pujehun'],
    'Males': [377, 153, 279],
    'Females': [207, 84, 146]
}

df = pd.DataFrame(data)

# Calculate total counts for males and females
total_males = df['Males'].sum()
total_females = df['Females'].sum()

# Prepare data for the pie chart
gender_data = [total_males, total_females]
gender_labels = ['Males', 'Females']

# Set the color palette
colors = sns.color_palette("bright")

# Create a pie chart with increased figure size
plt.figure(figsize=(8, 8))  # Increased figure size

# Define a function for formatting the label with increased font size
def label_format(percent):
    return f'{percent:.1f}%'

# Create the pie chart with increased label size
plt.pie(gender_data, labels=gender_labels, colors=colors, autopct=label_format, startangle=140, textprops={'fontsize': 16})

# Adjust title position
plt.title('Overall Gender Representation', fontsize=16, pad=20)  # Added pad to move title down

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```
