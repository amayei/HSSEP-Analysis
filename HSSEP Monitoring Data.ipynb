```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
try:
    df = pd.read_excel('Process_indicator.xlsx')
    print("Data loaded successfully!")
except FileNotFoundError:
    print("Error: The file 'Process_indicator.xlsx' was not found.")
    exit()

print("Data Overview:")
print(df.head())

# Check if required columns exist
output_columns = [col for col in df.columns if col.startswith('Output_') and col.endswith('_Activities')]
if not output_columns:
    print("Error: No columns found starting with 'Output_' and ending with '_Activities'")
    exit()

if 'total_females' not in df.columns or 'total_males' not in df.columns:
    print("Error: 'total_females' or 'total_males' columns are missing")
    exit()

# Prepare data for visualization
output_data = []

for col in output_columns:
    output_number = col.split('_')[1]
    total_females = df[df[col].notna()]['total_females'].sum()
    total_males = df[df[col].notna()]['total_males'].sum()
    output_data.append({
        'Output': f'Output {output_number}',
        'Females': total_females,
        'Males': total_males
    })

df_output = pd.DataFrame(output_data)

# Melt the dataframe for easier plotting
df_melted = pd.melt(df_output, id_vars=['Output'], var_name='Gender', value_name='Count')

# Create the visualization
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df_melted, x='Output', y='Count', hue='Gender', palette=['#FF69B4', '#87CEEB'])

# Adding labels to show exact counts on the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=10, color='black', 
                xytext=(0, 5), textcoords='offset points')

plt.title('Gender Involvement per Output', fontsize=16)
plt.xlabel('Outputs', fontsize=12)
plt.ylabel('Number of Participants', fontsize=12)
plt.legend(title='Gender')
plt.tight_layout()

# Save the plot as an image file
plt.savefig('gender_involvement_per_output.png')
print("Plot saved as 'gender_involvement_per_output.png'")

# Optional: Show the plot (uncomment if running in an interactive environment)
# plt.show()
```

```python
# Step 2: Extract unique activities from output 3 and 4
activities_out3 = df['Output_3_Activities'].dropna().unique()
activities_out4 = df['Output_4_Activities'].dropna().unique()

# Combine the unique activities from both outputs
unique_activities = set(activities_out3).union(set(activities_out4))

# Display the list of unique activities
print("\nList of Unique Activities Implemented:")
for i, activity in enumerate(unique_activities, 1):
    print(f"{i}. {activity}")

# Total number of unique activities
total_activities = len(unique_activities)
print(f"\nTotal Number of Unique Activities Implemented: {total_activities}")
```

```python
# Step 3: Summarize data by Output (Output 3 and Output 4)
output_summary = df.groupby('Output').agg({
    'Output_3_Activities': 'count',            # Count of activities for Output 3
    'Other_activity_Opt_3': 'count',           # Count of other activities for Output 3
    'Output_4_Activities': 'count',            # Count of activities for Output 4
    'Other_activity_Opt_4': 'count'            # Count of other activities for Output 4
}).reset_index()

# Display the summary
print("\nSummary by Output:")
print(output_summary)
```

```python
# Step 4: Summarize data by District
district_summary = df.groupby('District').agg({
    'Output_3_Activities': 'count',            # Count of activities for Output 3
    'Other_activity_Opt_3': 'count',           # Count of other activities for Output 3
    'Output_4_Activities': 'count',            # Count of activities for Output 4
    'Other_activity_Opt_4': 'count'            # Count of other activities for Output 4
}).reset_index()

# Display the summary
print("\nSummary by District:")
print(district_summary)
```

## Total Participants Per Output Activity

```python
import pandas as pd

# Import data into pandas dataframe
df = pd.read_excel('Process_indicator.xlsx')

# Step 1: Explode the activities into separate rows
# This will create separate rows for each activity in Output 3 and Output 4
df_exploded_3 = df[['Output_3_Activities', 'total_participants']].copy()
df_exploded_4 = df[['Output_4_Activities', 'total_participants']].copy()

# Use explode to split the lists into separate rows
df_exploded_3['Output_3_Activities'] = df_exploded_3['Output_3_Activities'].str.split(',')
df_exploded_4['Output_4_Activities'] = df_exploded_4['Output_4_Activities'].str.split(',')

df_exploded_3 = df_exploded_3.explode('Output_3_Activities').reset_index(drop=True)
df_exploded_4 = df_exploded_4.explode('Output_4_Activities').reset_index(drop=True)

# Step 2: Aggregate total participants by activity in Output 3 and Output 4
total_participants_output_3 = df_exploded_3.groupby('Output_3_Activities')['total_participants'].sum().reset_index()
total_participants_output_4 = df_exploded_4.groupby('Output_4_Activities')['total_participants'].sum().reset_index()

# Step 3: Rename columns for clarity
total_participants_output_3.columns = ['Activity', 'Total_Participants_Output_3']
total_participants_output_4.columns = ['Activity', 'Total_Participants_Output_4']

# Step 4: Display the total participants for each activity
print("Total Participants for Output 3 Activities:")
print(total_participants_output_3)

print("\nTotal Participants for Output 4 Activities:")
print(total_participants_output_4)
```
